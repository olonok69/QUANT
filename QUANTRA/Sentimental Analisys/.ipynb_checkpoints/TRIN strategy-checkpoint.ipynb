```python
def variance_calculator(series,series_average,win_len):
    sma=win_len
    temp=series.subtract(series_average)
    temp2= temp.apply(lambda x:x**2)
    temp3=temp2.rolling(sma-1).mean()
    sigma=temp3.apply(lambda x:math.sqrt(x))
    return sigma
```

```python
Data1 = quandl.get('CHRIS/CME_SP1',start_date="2014-12-12")
```

```python
declining = quandl.get('URC/NYSE_DEC',start_date="2014-12-12")
advancing= quandl.get('URC/NYSE_ADV',start_date="2014-12-12")
```

```python
adv_vol = quandl.get('URC/NYSE_ADV_VOL',start_date="2014-12-12")
dec_vol = quandl.get('URC/NYSE_DEC_VOL',start_date="2014-12-12")
```

```python
Data=declining
Data['declining']=declining['Numbers of Stocks']
Data['advancing']=advancing['Numbers of Stocks']
Data['dec_vol']=dec_vol['Numbers of Stocks']
Data['adv_vol']=adv_vol['Numbers of Stocks']
merged=Data.join(Data1)
```

```python
merged=merged.fillna(method='ffill')
```

```python
Data=merged
```

```python
tem1=Data['advancing'].divide(Data['declining'])
tem2=Data['adv_vol'].divide(Data['dec_vol'])
```

```python
tem=tem1.divide(tem2)
Data['TRIM']=tem
Data['TRIM']=Data['TRIM'].apply(lambda x:math.log(x))
Data['future']=Data['Last']
Data.to_csv("temp_data.csv")
Data=pd.read_csv("temp_data.csv")
```

```python
sma=22 # moving average window length
k=1.5 # constant representing k times sigma away from moving average (for bollinger bands)
l=2 # constant representing l times sigma away from bollinger band (for stoploss band )
pro=0 # profit variable
flag=1 # flag is there to begin first transaction --tarnsaction should start with LBB/UBB crossing only
buy_flag=False
sell_flag=False
transaction_start_price=0
abs_SL=25 # absolute stoploss variable
mtm=list()
order_details=list()
order=list() # order is a lisst which contains orders BUY /SELL /DO_Nothing
profit=list()
buy_sell=list()
stoploss=list()
trade_cause=list()

```

```python
Data['mAvg']=Data['TRIM'].rolling(sma).mean() # calculating the moving average of the trim
```

```python
Data['TRIM_Prev']=Data['TRIM'].shift(1)
Data.to_csv("temp_data.csv")
Data=pd.read_csv("temp_data.csv")
```

```python
Data.tail()
```

```python
sigma=variance_calculator(Data['TRIM'],Data['mAvg'],sma) # calculating the standar deviation
```

```python
k_sigma=k*sigma
l_sigma=l*sigma
Data['UBB']=Data['mAvg'].add(k_sigma) # upper bollinger band
Data['LBB']=Data['mAvg'].subtract(k_sigma) # upper bollinger band
Data['USL']=Data['UBB'].add(l_sigma) # upper stoploss band
Data['LSL']=Data['LBB'].subtract(l_sigma) # lower stoploss band
```

```python
Data['order']=pd.Series() # order is a lisst which contains orders BUY /SELL /DO_Nothing
```

```python
s=Data['TRIM'].size
```

```python
for i in range(s):
    pro=0
    future_cost=Data['future'][i]
    TRIN=Data['TRIM'][i]
    TRIN_prev=Data['TRIM_Prev'][i]
    LBB=Data['LBB'][i]
    UBB=Data['UBB'][i]
    mAvg=Data['mAvg'][i]
    USL=Data['USL'][i]
    LSL=Data['LSL'][i]
    UBB_cross=(TRIN>UBB) and(TRIN_prev<UBB) # variable TRIN crosses upper bollinger Band
    LBB_cross=(TRIN<LBB) and(TRIN_prev>LBB) # variable TRIN crosses lower bollinger Band
    mAvg_cross_up=(TRIN>mAvg) and(TRIN_prev<mAvg) # Check if TRIN crosses moving average from low to high
    mAvg_cross_down=(TRIN<mAvg) and(TRIN_prev>mAvg) # Check if TRIN crosses moving average from low to low
    USL_cross=(TRIN>USL) and(TRIN_prev<USL) # Check if TRIN crosses upper stoploss band
    LSL_cross=(TRIN<LSL) and(TRIN_prev>LSL) # Check if TRIN crosses upper stoploss band
    
    if(UBB_cross and(not buy_flag)and flag==1):
        flag=0
        buy_flag=True
        sell_flag=False
        transaction_start_price=future_cost
        order_details=[1,"Buy","UBB crossed","0","position taken"]
    elif(LBB_cross and(not sell_flag) and flag==1):
        flag=0
        buy_flag=False
        sell_flag=True
        transaction_start_price=future_cost
        order_details=[-1,"Sell","LBB crossed","0","position taken"]
    elif(mAvg_cross_up and(not buy_flag) and flag==0):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= future_cost -transaction_start_price
        order_details=[1,"Buy","mAvg crossed","0","position closed"]
    elif(LSL_cross and(not buy_flag) and flag==0):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= future_cost-transaction_start_price
        order_details=[1,"Buy","LSB crossed","stoploss executed","position closed"]
    elif((future_cost-transaction_start_price)> abs_SL and flag==0 and (not buy_flag)):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= future_cost-transaction_start_price
        order_details=[1,"Buy","LSB crossed","stoploss executed","position closed"]
    elif(mAvg_cross_down and(not sell_flag) and flag==0):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= -(future_cost -transaction_start_price)
        order_details=[-1,"Sell","mAvg crossed(h to l)","0","position closed"]
    elif(USL_cross and(not sell_flag) and flag==0):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= -(future_cost-transaction_start_price)
        order_details=[-1,"Sell","USB crossed","stoploss executed","position closed"]
    elif((-future_cost-transaction_start_price)> abs_SL and flag==0 and (not sell_flag)):
        flag=1
        buy_flag=False
        sell_flag=False
        pro= -(future_cost-transaction_start_price)
        order_details=[-1,"Sell","USB crossed","stoploss executed_abs","position closed"]
    else: #MTM each day
        if(buy_flag==0 and sell_flag==0): tempo="0"
        else:
            if(buy_flag==1 and sell_flag==0):tempo=(Data['future'][i]-transaction_start_price)*500
            if(buy_flag==0 and sell_flag==1):tempo=(-Data['future'][i]+transaction_start_price)*500
        order_details=[0,"No Trade", "no trade","0", tempo]
   
    profit.append(pro)
    order.append(order_details[0])
    buy_sell.append(order_details[1])
    trade_cause.append(order_details[2])
    stoploss.append(order_details[3])
    mtm.append(order_details[4])
```

```python
Data['placed_order']=pd.Series(order)
Data['cost']= -(Data['placed_order'].multiply(Data['future']))*500 # cost each transaction
Data['out']=Data['cost'].cumsum()# cummulative cost profit/lost after transaction after till now
Data['buy_sell']=pd.Series(buy_sell)
Data['profit']=-pd.Series(profit)*500
Data['stoploss']=pd.Series(stoploss)
Data['trade_cause']=pd.Series(trade_cause)
Data['mtm']=pd.Series(mtm)
print(Data['out'])
```

```python
output=pd.DataFrame()
output['date']=Data['Date']
output['Close']=Data['future']
output['TRIN']=Data['TRIM']
output['place_order']=Data['placed_order']
output['buy_sell']=Data['buy_sell']
output['trade_cause']=Data['trade_cause']
output['PnL']=Data['profit']
output['mtm']=Data['mtm']
output['stoploss']=Data['stoploss']
output['Cash Account']=Data['out']
output.to_excel('TRIN_SL_output.xlsx',sheet_name='Sheet1')
```
